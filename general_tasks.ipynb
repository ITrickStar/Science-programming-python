```python
# List divider
class list_divider(list):
    def __truediv__(self, other):
        if other <= 0:
            raise ValueError

        res = []
        div, mod = divmod(len(self), other)
        shift = 0

        for i in range(other):
            if mod:
                res.append(self[i * div + shift : (i + 1) * div + shift + 1])
                mod -= 1
                shift += 1
            else:
                res.append(self[i * div + shift : (i + 1) * div + shift])
        return res
```
